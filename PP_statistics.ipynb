""" Script for generating barplots for three parameters in dataset with predicted prepositional senses:
    - frequency distribution of prepositions found in dataset;
    - frequency distribution of syntaxemes (preposition + dependent case combinations) found in dataset;
    - frequency distribution of prepositional senses predicted.
"""

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Load dataset with predicted labels

```python
filename = 'oracles_predicted.csv'
df = pd.read_csv(filename)
```

```python
df
```

### Prepare 'value:count' dictionaries for prepositions, prepositional syntaxemes and labels

```python
prep_dict = dict(df['preposition'].value_counts()) # for prepositions
ft_prep_dict = dict(df['ft_prep'].value_counts()) # for prepositional syntaxemes
sense_dict = dict(df['label'].value_counts()) # for predicted senses
```

### Plot frequency distribution of prepositions

```python
total = len(df)
```

```python
filename_noext = filename[:-4] # used in graph filenames
```

```python
plt.figure(figsize = (10, 6))
plt.title(f'Частотное распределение предлогов (всего конструкций: {total})')
ax = plt.gca()
ax.set_facecolor('#eae9f3')
ax.tick_params(axis=u'both', which=u'both',length=0)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
bars = plt.bar(x = prep_dict.keys(), height = prep_dict.values(), color = '#ed2745', zorder = 3, align = 'center')
plt.grid(axis = 'y', color = 'white', zorder = 0)
plt.xticks(rotation = 45)
plt.margins(0.01, 0.05)
plt.tight_layout()

plt.savefig(f'{filename_noext}_barplot_prepositions.png', dpi = 300)
```

### Plot frequency distribution of syntaxemes

```python
plt.figure(figsize = (10, 6))
plt.title(f'Частотное распределение синтаксем (всего конструкций: {total})')
ax = plt.gca()
ax.set_facecolor('#eae9f3')
ax.tick_params(axis=u'both', which=u'both',length=0)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
bars = plt.bar(x = ft_prep_dict.keys(), height = ft_prep_dict.values(), color = '#ed2745', zorder = 3, align = 'center')
plt.grid(axis = 'y', color = 'white', zorder = 0)
plt.xticks(rotation = 45)
plt.margins(0.01, 0.05)
plt.tight_layout()

plt.savefig(f'{filename_noext}_barplot_syntaxemes', dpi = 300)
```

### Plot frequency distribution of labels

```python
plt.figure(figsize = (10, 6))
plt.title(f'Частотное распределение значений (всего конструкций: {total})')
ax = plt.gca()
ax.set_facecolor('#eae9f3')
ax.tick_params(axis=u'both', which=u'both',length=0)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
bars = plt.bar(x = sense_dict.keys(), height = sense_dict.values(), color = '#ed2745', zorder = 3, align = 'center')
plt.grid(axis = 'y', color = 'white', zorder = 0)
plt.xticks(rotation = 45)
plt.margins(0.01, 0.05)
plt.tight_layout()

plt.savefig(f'{filename_noext}_barplot_senses', dpi = 300)
```
